### LeadLag_RT()

```python
help(LeadLag_RT)
```

```python
TSim = 100
Ts = 0.1
N = int(TSim/Ts) + 1

# Path for MV
MVPath = {0: 0, 5: 1, 50: 2, 80: 3, TSim: 3}

t = []
MV = []
MVDelay = []
PV_EBD = []
PV_EFD = []
PV_TRAP = []

# Parameters
Kp = 1
Tlead = 5
Tlag = 10
theta = 10

for i in range(0,N):
    t.append(i*Ts)
    SelectPath_RT(MVPath,t,MV)
    Delay_RT(MV,theta,Ts,MVDelay)
    LeadLag_RT(MVDelay,Kp,Tlead,Tlag,Ts,PV_EBD,0,'EBD')
    LeadLag_RT(MVDelay,Kp,Tlead,Tlag,Ts,PV_EFD,0,'EFD')
    LeadLag_RT(MVDelay,Kp,Tlead,Tlag,Ts,PV_TRAP,0,'TRAP')
    
plt.figure(figsize = (15,9))

plt.subplot(4,1,1)
plt.step(t,MV,'b-',label='MV',where='post')
plt.step(t,MVDelay,'b-',linewidth=0.5,label='MV plus delay',where='post')
plt.ylabel('Value of MV')
plt.title('First order plus delay response')
plt.legend(loc='best')
plt.xlim([0, TSim])

plt.subplot(4,1,2)
plt.step(t,PV_EBD,'g-',label='First order response plus delay (EBD)',where='post')
plt.ylabel('Value of PV')
plt.xlabel('Time [s]')
plt.legend(loc='best')
plt.xlim([0, TSim])

plt.subplot(4,1,3)
plt.step(t,PV_EFD,'m-',label='First order response plus delay (EFD)',where='post')
plt.ylabel('Value of PV')
plt.xlabel('Time [s]')
plt.legend(loc='best')
plt.xlim([0, TSim])  

plt.subplot(4,1,4)
plt.step(t,PV_TRAP,'r-',label='First order response plus delay (TRAP)',where='post')
plt.ylabel('Value of PV')
plt.xlabel('Time [s]')
plt.legend(loc='best')
plt.xlim([0, TSim]) 
```

### LeadLag() with widget

```python
def leadLagPlot(Kp, Tlead, Tlag, theta, Ts):
  TSim = 100
  N = int(TSim/Ts) + 1

  t = []
  MV = []
  MVDelay = []
  PV_EBD = []
  PV_EFD = []
  PV_TRAP = []

  MVPath = {0: 0, 5: 1, 50: 2, 80: 3, TSim: 3}

  for i in range(0,N):
    t.append(i*Ts)
    SelectPath_RT(MVPath,t,MV)
    Delay_RT(MV,theta,Ts,MVDelay)
    LeadLag_RT(MVDelay,Kp,Tlead,Tlag,Ts,PV_EBD,0,'EBD')
    LeadLag_RT(MVDelay,Kp,Tlead,Tlag,Ts,PV_EFD,0,'EFD')
    LeadLag_RT(MVDelay,Kp,Tlead,Tlag,Ts,PV_TRAP,0,'TRAP')
 

  plt.figure(figsize = (15,9))

  plt.subplot(2,1,1)
  plt.step(t,MV,'b-',label='MV',where='post')
  plt.step(t,MVDelay,'b-',linewidth=0.5,label='MV plus delay',where='post')
  plt.ylabel('Value of MV [%]')
  plt.title('First order plus delay response')
  plt.legend(loc='best')
  plt.xlim([0, TSim])

  plt.subplot(2,1,2)
  plt.step(t,PV_EBD,'g-',label='First order response plus delay (EBD)',where='post')
  plt.step(t,PV_EFD,'lime',label='First order response plus delay (EFD)',where='post')
  plt.step(t,PV_TRAP,'r-',label='First order response plus delay (TRAP)',where='post')
  plt.ylabel('Value of PV [¬∞C]')
  plt.xlabel('Time [s]')
  plt.legend(loc='best')
  plt.xlim([0, TSim]) 

interact(leadLagPlot, Kp = widgets.FloatSlider(value=1,
                                                min=0.1,
                                                max=1.0,
                                                step=0.1),
                    Tlead = widgets.FloatSlider(value=15,
                                                min=-5,
                                                max=15.0,
                                                step=1),
                    Tlag = widgets.FloatSlider(value=10,
                                                min=10,
                                                max=30,
                                                step=5),
                    theta = widgets.FloatSlider(value=10,
                                                min=1,
                                                max=20.0,
                                                step=1),
                    Ts = widgets.FloatSlider(value=0.1,
                                                min=0.1,
                                                max=1,
                                                step=0.1))
```

## PID_RT()

```python
help(PID_RT)
```

```python
# Path for SP, PV, Man, MVMan, MVFF

SPPath =  {0: 0, 25 : 50}
PVPath = {0: 0, 25: 45}
ManPath = {0: False, 450: True, 500: False}
MVManPath = {0: 0, 450: 50}
ManFFPath = {0: False, 800: True}
MVFFPath = {0:0, 800:10}


TSim =1000
Ts = 1
N = int(TSim/Ts) + 1


t = []

SP = []
PV = []
Man = []
MVMan = []
ManFF = []
MV = []
MVP = []
MVI = []
MVD = []
MVFF = []
E = []

#PID parameters
Kc = 6.00 
Ti = 160.00
Td = 3.50

alpha = 1

PVinit = 0
MVMin = 0
MVMax = 100


for i in range(0,N):
    t.append(i*Ts)
    SelectPath_RT(SPPath,t,SP)
    SelectPath_RT(PVPath,t,PV)  
    SelectPath_RT(ManPath,t,Man)
    SelectPath_RT(MVManPath,t,MVMan)  
    SelectPath_RT(ManFFPath,t,ManFF)
    SelectPath_RT(MVFFPath,t,MVFF)
    PID_RT(SP, PV, Man, MVMan, MVFF, Kc, Ti, Td, alpha, Ts, MVMin, MVMax, MV, MVP, MVI, MVD, E, ManFF, PVinit, method='EBD-EBD')
    
plt.figure(figsize = (25,15))

plt.subplot(4,1,4)
plt.step(t,Man,'magenta',label='Man', linewidth = 2, where='post')
plt.ylabel('Value of Man [0/1]')
plt.legend(loc='upper right')
plt.xlim([0, TSim])
plt.ylim([-0.1,1.1])
plt.xlabel('Time [s]')

plt.subplot(4,1,2)
plt.step(t,E,'olive',label='E', linewidth = 2, where='post')
plt.ylabel('Value of E [¬∞C]')
plt.legend(loc='upper right')
plt.xlim([0, TSim])

plt.subplot(4,1,3)
plt.step(t,SP,'black',label='SP', linewidth = 2, where='post')
plt.ylabel('Value of SP [¬∞C]')
plt.legend(loc='upper right')
plt.xlim([0, TSim])

plt.subplot(4,1,1)
plt.step(t,MV,'blue',label='MV', linewidth = 2,where='post')
plt.step(t,MVP,'green',label='MVP',  linewidth = 1, where='post')
plt.step(t,MVI,'orange',label='MVI', linewidth = 1, where='post')
plt.step(t,MVD,'purple',label='MVD',dashes=[3,3], linewidth = 1, where='post')
plt.step(MVFF, 'cyan' , label= 'MVFF',linewidth = 1, where='post')
plt.title('Step response of the PID Controller')
plt.ylabel('Value of MV [%]')
plt.legend(loc='upper right')
plt.xlim([0, TSim])
```
